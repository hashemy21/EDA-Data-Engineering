```python
tmp=athlete_events[athlete_events['Season'] == 'Summer']
tmp2=athlete_events[athlete_events['Season'] == 'Winter']

df3=tmp.groupby('Season')['ID'].count()
df4=tmp2.groupby('Season')['ID'].count()

# merged=pd.concat(df3,df4)


summer=df3['Summer']
winter=df4['Winter']

season_fig = plt.figure(figsize=(5, 8), dpi=100)
# ax = season_fig.add_subplot(1,1,1)
# x = np.arange(1,len(df3)+1)
# width = 0.2
# ax.bar(height= df3 , width=width,  x = x-0.3/2)
# ax.bar(height =df4,width=width, x = x+0.3/2)
# ax.set_xticklabels(df3.index,fontsize=7)
# ax.set_xticks(x)

objects = ('Summer', 'Winter')
y_pos = np.arange(len(objects))
performance = [summer,winter]

plt.bar(y_pos, performance, align='center', alpha=0.5)
plt.xticks(y_pos, objects)
plt.ylabel('participants')
plt.title('Summer/Winter participants')
```

### This graph is showing the Summer events compared to the Winter events with respect to the number of the participants, and the Summer events was larger than the Winter events

```python
plt.figure(figsize=(15, 6), dpi=85)

new1=athlete_events[athlete_events['Sport'] == 'Basketball']
df1=new1.groupby('Year')['Sport'].count()

new2=athlete_events[athlete_events['Sport'] == 'Tug-Of-War']
df2=new2.groupby('Year')['Sport'].count()

new3=athlete_events[athlete_events['Sport'] == 'Rugby']
df3=new3.groupby('Year')['Sport'].count()

new4=athlete_events[athlete_events['Sport'] == 'Table Tennis']
df4=new4.groupby('Year')['Sport'].count()

new5=athlete_events[athlete_events['Sport'] == 'Swimming']
df5=new5.groupby('Year')['Sport'].count()

new6=athlete_events[athlete_events['Sport'] == 'Athletics']
df6=new6.groupby('Year')['Sport'].count()

new7=athlete_events[athlete_events['Sport'] == 'Tennis']
df7=new7.groupby('Year')['Sport'].count()

new8=athlete_events[athlete_events['Sport'] == 'Triathlon']
df8=new8.groupby('Year')['Sport'].count()

plt.plot(df1.index,df1.values, label="Basketball")
plt.plot(df2.index,df2.values, label="Tug-Of-War")
plt.plot(df3.index,df3.values, label="Rugby")
plt.plot(df4.index,df4.values, label="Table Tennis")
plt.plot(df5.index,df5.values, label="Swimming")
plt.plot(df6.index,df6.values, label="Athletics")
plt.plot(df7.index,df7.values, label="Tennis")
plt.plot(df8.index,df8.values, label="Triathlon")

plt.title('Evolution of Sports over time')
plt.legend()
```

### This graph is showing the evolution of the sports over the time, and here is we see that some sports remain growing uniformly, like "Athletics" and "Swimming ", and some sports vanished in 1920 like Tug-Of-War, and some sports are new to the Olyimpic games like "Triathlon" that has begun particpating in 2000
